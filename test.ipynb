```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(test_size=0.2,random_state=42)
for train_index,test_index in split.split(housing,housing['chas']):
    strat_train_set=housing.loc[train_index]
    strat_test_set=housing.loc[test_index]
```

```python
housing=strat_train_set.drop("medv",axis=1)
housing_labels=strat_train_set["medv"].copy()
```

```python
housing_test=strat_test_set.drop("medv",axis=1)
housing_labels_test=strat_test_set["medv"].copy()
```

# MISSING DATA

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(housing)
```

```python
imputer.statistics_
```

```python
X=imputer.transform(housing)
```

```python
housing_tr=p.DataFrame(X,columns=housing.columns)
```

```python
housing_tr.describe()
```

# SELECTING A MODEL

```python
from sklearn.linear_model import LinearRegression
```

# USING DECISION TREE

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = LinearRegression()
model.fit(housing_tr,housing_labels)
```

# EVALUATING THE MODEL (RMSE)

```python
import numpy as np
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_test)
mse = mean_squared_error(housing_labels_test,housing_predictions)
rmse = np.sqrt(mse)
```
